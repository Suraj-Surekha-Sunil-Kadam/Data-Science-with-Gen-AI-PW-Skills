##Question 1:  What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

**Answer:**


What is Deep Learning?
- Definition: Deep Learning is a subset of machine learning that uses artificial neural networks with many layers (hence “deep”) to model complex relationships in data.
- Core Idea: Instead of manually designing features, deep learning models learn hierarchical representations directly from raw data (e.g., pixels in images, audio signals, text).

---

Evolution of Deep Learning
- 1950s-1980s: Early neural network concepts (Perceptron, backpropagation) laid the foundation.
- 1990s-2000s: Traditional ML dominated with algorithms like decision trees, support vector machines, and logistic regression due to limited computing power.
- 2010s onward: Explosion of big data, GPUs, and improved algorithms (e.g., convolutional neural networks, recurrent networks) enabled deep learning to outperform traditional ML in vision, speech, and natural language tasks.
- Today: Deep learning powers applications like self-driving cars, medical imaging, chatbots, and recommendation systems.

# Key Differences Between Deep Learning and Traditional Machine Learning

- **With Deep Learning:** Features are learned automatically from raw data  
  **While with Traditional ML:** Features must be manually engineered by experts  

- **With Deep Learning:** Requires large datasets to achieve high performance  
  **While with Traditional ML:** Works well even with smaller datasets  

- **With Deep Learning:** Needs high computational power (GPUs/TPUs) for training  
  **While with Traditional ML:** Can run efficiently on standard CPUs  

- **With Deep Learning:** Excels at handling unstructured data (images, audio, text)  
  **While with Traditional ML:** Performs best on structured/tabular data  

- **With Deep Learning:** Training is resource-intensive and time-consuming  
  **While with Traditional ML:** Training is faster and less resource-heavy  

- **With Deep Learning:** Models are often considered “black boxes” (less interpretable)  
  **While with Traditional ML:** Models are easier to interpret and explain  

# Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?

**Answer:**

## Basic Architecture of a Perceptron
- A **Perceptron** is the simplest type of artificial neural network, introduced by Frank Rosenblatt in 1958.  
- It consists of:
  - **Input layer:** Accepts multiple input features (x₁, x₂, …, xₙ).  
  - **Weights (w₁, w₂, …, wₙ):** Each input is multiplied by a weight that represents its importance.  
  - **Summation function:** Computes the weighted sum of inputs:  
    \[
    z = \sum (w_i \cdot x_i) + b
    \]
    where *b* is the bias term.  
  - **Activation function:** Applies a threshold (step function) to decide the output (0 or 1).  

---

## Functioning of a Perceptron
1. **Input stage:** The perceptron receives input values.  
2. **Weighted sum:** Each input is multiplied by its weight and added together with bias.  
3. **Activation:** The summation result is passed through an activation function (commonly a step function).  
4. **Output:** Produces a binary output (e.g., 0 or 1) depending on whether the weighted sum crosses the threshold.  

Example:  
- If \( z > 0 \), output = 1  
- If \( z \leq 0 \), output = 0  

---

## Limitations of a Perceptron
- **Linear separability:** Can only solve problems where data is linearly separable (e.g., AND, OR).  
- **Fails on complex tasks:** Cannot solve non-linear problems like XOR.  
- **Limited learning capacity:** Single-layer perceptrons cannot learn hierarchical or complex patterns.  
- **Binary output only:** Original perceptron uses a step function, restricting it to binary classification.  
- **No hidden layers:** Lacks the depth needed to capture complex relationships in data.  

---

## Summary
The perceptron is the **foundation of neural networks**, demonstrating how inputs, weights, and activation functions combine to produce outputs. However, its inability to handle non-linear problems led to the development of **multi-layer perceptrons (MLPs)** and modern deep learning architectures.

# Question 3: Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.  
**Answer:**

## Purpose of Activation Function
- Activation functions introduce **non-linearity** into neural networks, allowing them to learn complex patterns.  
- Without activation functions, a neural network would behave like a linear regression model, unable to capture intricate relationships.  
- They help determine whether a neuron should be activated (i.e., pass its signal forward) based on the weighted input.

---

## Comparison of Sigmoid, ReLU, and Tanh Functions

- **With Sigmoid:** Output is between 0 and 1  
  **While with ReLU:** Output is 0 for negative inputs and linear for positive inputs  
  **While with Tanh:** Output ranges from -1 to 1  

- **With Sigmoid:** Good for probabilistic interpretation  
  **While with ReLU:** Good for fast convergence and sparse activation  
  **While with Tanh:** Good for centered outputs around zero  

- **With Sigmoid:** Suffers from vanishing gradient problem for large inputs  
  **While with ReLU:** Can suffer from dying neurons (outputs stuck at 0)  
  **While with Tanh:** Also suffers from vanishing gradients but less than sigmoid  

- **With Sigmoid:** Used in binary classification tasks  
  **While with ReLU:** Commonly used in hidden layers of deep networks  
  **While with Tanh:** Used in hidden layers when zero-centered output is preferred  

---

## Summary
Activation functions are essential for enabling neural networks to learn **non-linear mappings**. While **Sigmoid** is useful for output layers in classification, **ReLU** is preferred for hidden layers due to its simplicity and efficiency, and **Tanh** offers a middle ground with zero-centered outputs but still faces gradient issues.

# Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.
**Answer:**

## Loss Function
- **Definition:** A loss function measures the error for a **single training example**.  
- **Purpose:** It tells us how far the predicted output is from the actual target for one data point.  
- **Examples:**
  - Mean Squared Error (MSE) for regression tasks  
  - Cross-Entropy Loss for classification tasks  

---

## Cost Function
- **Definition:** A cost function is the **average of the loss function across the entire training dataset**.  
- **Purpose:** It provides a global measure of how well the model is performing overall.  
- **Examples:**
  - Average MSE across all samples in regression  
  - Average Cross-Entropy Loss across all samples in classification  

---

## Key Difference

- **With Loss Function:** Error is calculated for a single training example  
  **While with Cost Function:** Error is averaged over the whole dataset  

- **With Loss Function:** Used to evaluate prediction quality at the individual sample level  
  **While with Cost Function:** Used to guide optimization of the entire model  

- **With Loss Function:** Examples include squared error for one data point or log loss for one sample  
  **While with Cost Function:** Examples include mean squared error (MSE) or average cross-entropy across all samples  

---

## Summary
The **loss function** focuses on the error of a single instance, while the **cost function** aggregates those errors to evaluate the model’s performance on the entire dataset. Both are crucial: loss functions define the error metric, and cost functions drive the optimization process during training.

# Question 5: Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.  
**Answer:**

## Role of Optimizers
- Optimizers are algorithms used to **update the weights and biases** of a neural network during training.  
- Their purpose is to **minimize the cost function** (overall error) by adjusting parameters in the right direction.  
- They determine **how fast and effectively** a model learns by controlling the step size (learning rate) and update rules.  

---

## Comparison of Gradient Descent, Adam, and RMSprop

- **With Gradient Descent:** Updates weights by moving in the direction of the negative gradient with a fixed learning rate  
  **While with Adam:** Combines momentum and adaptive learning rates for faster convergence  
  **While with RMSprop:** Adjusts learning rate based on recent gradients to handle non-stationary problems  

- **With Gradient Descent:** Simple and easy to implement, but can be slow and sensitive to learning rate choice  
  **While with Adam:** More efficient, works well with sparse gradients, and requires less tuning  
  **While with RMSprop:** Good for recurrent neural networks and stabilizes training by normalizing gradient updates  

- **With Gradient Descent:** May get stuck in local minima or oscillate if learning rate is not tuned properly  
  **While with Adam:** Handles noisy data and converges faster in practice  
  **While with RMSprop:** Prevents exploding or vanishing updates by scaling learning rates dynamically  

---

## Summary
Optimizers are crucial for guiding neural networks toward **better solutions**.  
- **Gradient Descent** is the foundational method, simple but limited.  
- **Adam** is widely used due to its efficiency and adaptability.  
- **RMSprop** is particularly effective for sequential data and recurrent networks.  
Together, they highlight how different strategies improve training stability and speed.

## Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


# AND gate truth table
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])   # inputs
y = np.array([0,0,0,1])  # outputs

# Initialize weights and bias
weights = np.random.rand(2)
bias = np.random.rand(1)
learning_rate = 0.1

# Activation function
def step_function(z):
    return 1 if z >= 0 else 0

# Training the perceptron
epochs = 20
errors = []

for epoch in range(epochs):
    total_error = 0
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = step_function(z)
        error = y[i] - y_pred
        # Update rule
        weights = weights + learning_rate * error * X[i]
        bias = bias + learning_rate * error
        total_error += abs(error)
    errors.append(total_error)
    print(f"Epoch {epoch+1}, Total Error: {total_error}")

print("\nFinal Weights:", weights)
print("Final Bias:", bias)

# Test the perceptron
print("\nTesting Perceptron:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    y_pred = step_function(z)
    print(f"Input: {X[i]} -> Output: {y_pred}")

# Plot training error
plt.plot(range(1, epochs+1), errors, marker='o')
plt.title("Training Error over Epochs (NumPy Perceptron)")
plt.xlabel("Epochs")
plt.ylabel("Total Error")
plt.show()


#Using TensorFlow/Keras

# Build a simple perceptron model
model = Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))  # single neuron

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
history = model.fit(X, y, epochs=100, verbose=0)

# Evaluate model
loss, acc = model.evaluate(X, y, verbose=0)
print("\nKeras Model Accuracy:", acc)

# Predictions
print("\nKeras Predictions:")
print(model.predict(X))

# Plot training accuracy
plt.plot(history.history['accuracy'])
plt.title("Training Accuracy (Keras Perceptron)")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```

## Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Generate input values
x = np.linspace(-10, 10, 200)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plot all activation functions
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(3, 1, 1)
plt.plot(x, y_sigmoid, color='blue')
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid(True)

# ReLU
plt.subplot(3, 1, 2)
plt.plot(x, y_relu, color='green')
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True)

# Tanh
plt.subplot(3, 1, 3)
plt.plot(x, y_tanh, color='red')
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

**Answer**

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical


# Load and preprocess MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values (0-255 -> 0-1)
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)


# Build the neural network model
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))        # Flatten 28x28 images into 784 vector
model.add(Dense(128, activation='relu'))        # Hidden layer with ReLU
model.add(Dense(64, activation='relu'))         # Another hidden layer
model.add(Dense(10, activation='softmax'))      # Output layer (10 classes)


# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


# Train the model
history = model.fit(X_train, y_train,
                    epochs=10,
                    batch_size=128,
                    verbose=1)


# Print training accuracy
train_acc = history.history['accuracy'][-1]
print(f"\nFinal Training Accuracy: {train_acc:.4f}")


# Plot training accuracy over epochs
plt.plot(history.history['accuracy'], marker='o')
plt.title("Training Accuracy over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```

## Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical


# Load and preprocess Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize pixel values (0-255 -> 0-1)
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)


# Build the neural network model
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))        # Flatten 28x28 images
model.add(Dense(128, activation='relu'))        # Hidden layer
model.add(Dense(64, activation='relu'))         # Another hidden layer
model.add(Dense(10, activation='softmax'))      # Output layer (10 classes)


# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


# Train the model
history = model.fit(X_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_data=(X_test, y_test),
                    verbose=1)


# Plot Loss and Accuracy Curves
plt.figure(figsize=(12,5))

# Loss curve
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# Accuracy curve
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title("Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Question 10: You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.
##● How would you design a deep learning model (perceptron or multilayer NN)?
##● Which activation function and loss function would you use, and why?
##● How would you train and evaluate the model, considering class imbalance?
##● Which optimizer would be suitable, and how would you prevent overfitting?


**Answer**

Problem framing and data pipeline
- Objective: Binary classification-flag transactions as fraudulent or legitimate with emphasis on recall (catch fraud) while controlling false positives (precision).
- Data ingestion (real-time):
- Streaming layer: Kafka/HTTP → feature service (low-latency transforms).
- Feature store: Consistent features for training/serving (e.g., transaction amount, merchant ID, customer location, device fingerprint, time features).
- Preprocessing:
- Numerical features: Standardization/robust scaling (amount, velocity features).
- Categorical features: Embeddings (merchant ID, customer location, device type) or target encoding with leakage-safe folds.
- Temporal features: Time-of-day, day-of-week, rolling aggregates (e.g., last 1h/24h spend).
- Data splits: Time-aware split (train/val/test by time) to respect production drift.

Model design (perceptron vs multilayer neural network)
- With Perceptron: Single linear decision boundary—too limited for complex fraud patterns.
While with Multilayer NN: Multiple hidden layers capture non-linear interactions (e.g., amount x merchant x location), better suited for fraud.
- Recommended architecture (tabular NN):
- Inputs:
- Numerical block: Dense → BatchNorm → ReLU.
- Categorical block: Embedding layers per high-cardinality feature (merchant ID, customer location).
- Fusion: Concatenate numerical and embedding outputs.
- Hidden stack: 2-4 Dense layers (e.g., 256→128→64), each with BatchNorm, ReLU, Dropout (0.2-0.5), L2 regularization.
- Output: Dense(1, activation=sigmoid) for fraud probability.

Activation and loss functions
- With Activation (hidden): ReLU for speed and stable gradients; consider LeakyReLU to avoid dying neurons.
While Output Activation: Sigmoid to produce calibrated probabilities in binary classification.
- With Loss Function:
- Weighted Binary Cross-Entropy using class weights to counter imbalance.
- Alternatively: Focal Loss to focus learning on hard, minority-class examples (tunable γ, α).
Why: Fraud is rare—standard BCE underweights minority errors; weighting/focal loss improves recall without exploding false positives.

Training and evaluation under class imbalance
- Sampling strategy:
- With Class Weights: Compute weights from label distribution (e.g., fraud gets higher weight).
While with Resampling: Use undersampling of majority or SMOTE for offline experiments—avoid SMOTE for production due to synthetic artifacts.
- Metrics:
- Primary: Precision-Recall AUC, Recall@fixed precision, F1, False Positive Rate at operational threshold.
- Secondary: ROC AUC, Calibration (reliability curves).
- Threshold tuning: Optimize threshold for business cost (fraud miss cost vs investigation cost).
- Validation: Time-split validation; cross-validate with grouped folds (by customer/merchant) to reduce leakage.
- Evaluation protocol:
- Confusion matrix at chosen threshold.
- Cost-sensitive analysis: Expected savings vs manual review load.
- Drift checks: Monitor feature distributions and performance over time.

Optimizer and overfitting prevention
- With Optimizer: Adam (adaptive learning rates, fast convergence) with learning rate scheduling (ReduceLROnPlateau).
While Alternatives: RMSprop for non-stationary gradients; SGD+Momentum if you need stricter regularization and smoother convergence.
- Regularization & stability:
- Dropout: 0.2-0.5 in hidden layers.
- L2 weight decay: e.g., 1e-4-1e-5.
- BatchNorm: Stabilizes training and helps generalization.
- Early stopping: Monitor val PR-AUC with patience (e.g., 5-10 epochs).
- Data augmentation (tabular): Careful noise injection to numeric features; avoid unrealistic synthetic patterns.

Real-time serving and monitoring
- Inference:
- Low-latency API (TensorFlow Serving/ONNX Runtime) with feature parity to training.
- Thresholding: Dynamic thresholds by segment (merchant/customer risk tiers).
- Actioning: Route high-risk transactions to manual review; auto-block extreme risk.
- Monitoring:
- Live metrics: Precision/recall drift, alert rates, investigation queue load.
- Calibration: Periodic recalibration (Platt scaling/Isotonic) if score distributions shift.
- Retraining cadence: Weekly/monthly with fresh labels, time-aware splits, and shadow deployments before full rollout.
