# Question 1: What is the role of filters and feature maps in Convolutional Neural Network (CNN)?

**Answer:**

**Filters** (also called kernels)
- Definition: Filters are small matrices (e.g., 3×3 or 5×5) that slide across the input image during convolution.
- Role:
- Detect local patterns such as edges, corners, textures, or color gradients.
- Each filter learns a specific type of feature during training.
- Multiple filters allow the CNN to capture diverse aspects of the image (e.g., one filter may detect vertical edges, another horizontal edges).
- Learning: The values inside filters are not fixed; they are optimized during training through backpropagation.

**Feature Maps**
- Definition: A feature map is the output produced when a filter is applied to the input image.
- Role:
- Represents the presence and location of a learned feature in the image.
- Highlights regions where the filter detects strong patterns (e.g., a bright spot in the feature map indicates a strong match).
- Multiple feature maps are generated in each convolutional layer, corresponding to different filters.
- Hierarchy:
- Early layers produce feature maps for simple features (edges, lines).
- Deeper layers produce feature maps for complex features (shapes, objects).

**Relationship between Filters and Feature Maps**
- Filters are the learned detectors.
- Feature maps are the visual evidence of what those detectors find in the input.
- Together, they enable CNNs to automatically learn and represent hierarchical features, moving from low-level pixel patterns to high-level semantic concepts.

**Example (Cats vs Dogs Classifier)**
- A filter in the first layer might detect fur texture.
- The resulting feature map highlights regions of the image where fur patterns are present.
- Deeper filters might detect ears, eyes, or snouts, and their feature maps show where those features occur.
- Ultimately, the CNN combines these feature maps to decide whether the image is a cat or a dog.

# Question 2: Explain the concepts of padding and stride in CNNs(Convolutional Neural Network). How do they affect the output dimensions of feature maps?
**Answer:**

**Padding**
- Definition: Padding refers to adding extra pixels (usually zeros) around the input image before applying convolution.
- Purpose:
- Prevents loss of information at the edges of the image.
- Helps control the spatial size of the output feature maps.
- Ensures that filters can slide over the entire image, including borders.
- Types:
- Valid Padding: No padding is added. The output feature map is smaller than the input because the filter cannot go beyond the edges.
- Same Padding: Enough padding is added so that the output feature map has the same spatial dimensions as the input.

**Stride**
- Definition: Stride is the number of pixels by which the filter moves/slides across the input image during convolution.
- Purpose:
- Controls how much the filter shifts at each step.
- Larger strides reduce the size of the output feature map (downsampling).
- Smaller strides (e.g., stride = 1) preserve more spatial detail but produce larger feature maps.

**Effect on Output Dimensions**
The output size of a feature map depends on input size (W), filter size (F), padding (P), and stride (S). The formula is:
\mathrm{Output\  dimension}=\frac{W-F+2P}{S}+1
- Padding increases output size by effectively enlarging the input.
- Stride decreases output size by skipping positions as the filter slides.
- Choosing padding and stride carefully balances feature preservation vs. computational efficiency.

**Example**
- Input image: 32x32
- Filter size: 3x3
- Padding: 1 (same padding)
- Stride: 1
\mathrm{Output\  size}=\frac{32-3+2(1)}{1}+1=32
So the output feature map remains 32\times 32.
If stride = 2 (with same padding):
\mathrm{Output\  size}=\frac{32-3+2(1)}{2}+1=16
The feature map shrinks to 16\times 16.

# Question 3: Define receptive field in the context of CNNs. Why is it important for deep architectures?
**Answer:**

**Receptive Field**
  
-  Definition - The receptive field in a Convolutional Neural Network (CNN) refers to the region of the input image that a particular neuron (or unit) in a feature map is sensitive to.
- In simpler terms, it is the portion of the input image that influences the output of a specific neuron.
- For example, in the first convolutional layer, a neuron might only "see" a small patch (like 3\times 3 pixels). In deeper layers, neurons have larger receptive fields because they combine information from multiple earlier layers.

**How Receptive Field Grows**
- Early layers: Small receptive fields capture fine details such as edges, textures, or simple shapes.
- Deeper layers: As convolution and pooling operations stack, the receptive field expands. Neurons begin to capture more complex and global features, such as object parts or entire objects.
- This hierarchical growth allows CNNs to move from low-level to high-level feature representation.

**Importance in Deep Architectures**

- Capturing Context
Larger receptive fields allow neurons to consider more of the image context, which is crucial for recognizing complex patterns (e.g., distinguishing a cat's ear from background noise).
- Hierarchical Feature Learning
Deep CNNs rely on progressively larger receptive fields to build feature hierarchies — from edges → shapes → objects.
- Balance Between Detail and Abstraction
Small receptive fields preserve fine-grained details, while large receptive fields capture global structures. Both are necessary for accurate classification.
- Design Considerations
- If receptive fields are too small, the network may miss global context.
- If they are too large too early, the network may lose fine details.
- Proper architecture design ensures receptive fields grow appropriately across layers.

**Example**

In a cat vs dog classifier:
- A neuron in the first layer might detect fur texture (small receptive field).
- A neuron in deeper layers might detect the shape of an ear or snout (larger receptive field).
- Eventually, the network combines these features to classify the entire animal correctly.

# Question 4: Discuss how filter size and stride influence the number of parameters in a CNN.
**Answer:**

**Filter Size**
- Definition: The filter (or kernel) size refers to the dimensions of the matrix used in convolution (e.g., 3x3, 5x5).
- Impact on Parameters:
- Each filter has its own set of trainable weights.
- For a filter of size F\times F applied to an input with C channels, the number of parameters is:
\mathrm{Parameters\  per\  filter}=F\cdot F\cdot C+1- (the +1 accounts for the bias term).
- Larger filters mean more parameters, which increases model capacity but also computational cost and risk of overfitting.
- Smaller filters reduce parameters and computation, but may capture less spatial context.

**Stride**
- Definition: Stride is the step size with which the filter moves across the input image.
- Impact on Parameters:
- Stride does not change the number of parameters in the filter itself (the weights remain the same).
- Instead, stride affects the output feature map size:
- Larger stride → fewer positions → smaller feature maps.
- Smaller stride → more positions → larger feature maps.
- While stride doesn't alter the parameter count, it influences the computational load and the amount of information captured.

**Combined Effect**
- Filter size directly determines the number of trainable parameters.
- Stride indirectly influences the model's complexity by controlling the resolution of feature maps and the number of activations processed.
- Together, they balance model capacity, computational efficiency, and feature extraction detail.

**Example**
- Input: 128x128x3 image
- Filter size: 3x3, channels = 3
\mathrm{Parameters\  per\  filter}=3\cdot 3\cdot 3+1=28
- With 32 filters, total parameters = 32\cdot 28=896.
- Stride = 1 → large feature maps (high detail).
- Stride = 2 → smaller feature maps (less detail, faster computation).

# Question 5: Compare and contrast different CNN-based architectures like LeNet, AlexNet, and VGG in terms of depth, filter sizes, and performance.
**Answer**

**LeNet**
- Depth: Shallow network (~5-7 layers).
- Filter Sizes: Larger filters (e.g., 5x5) in early layers.
- Performance:
- Designed for digit recognition (MNIST dataset).
- Low computational requirements, suitable for early hardware.
- Pioneered the use of convolution + pooling + fully connected layers.
- Limitations: Too shallow for complex datasets like ImageNet; struggles with large-scale, high-resolution images.

**AlexNet**
- Depth: Much deeper (~8 layers: 5 convolutional + 3 fully connected).
- Filter Sizes:
- First layer used large filters (11x11) with stride 4.
- Later layers used smaller filters (3x3).
- Performance:
- Won the ImageNet Large Scale Visual Recognition Challenge (ILSVRC) 2012 by a huge margin.
- Introduced ReLU activation, Dropout, and GPU training for efficiency.
- Achieved ~15% top-5 error rate, far better than traditional methods at the time.
- Impact: Sparked the deep learning revolution in computer vision.


**VGG**
- Depth: Very deep (16-19 layers).
- Filter Sizes:
- Standardized to small filters (3x3) throughout the network.
- Used multiple stacked small filters instead of large ones, increasing non-linearity and representational power.
- Performance:
- Achieved strong results on ImageNet with ~7% top-5 error rate.
- Simple and elegant design, widely adopted as a backbone in transfer learning.
- Limitations: Computationally heavy, large number of parameters (~138 million), slower training compared to AlexNet.

# Question 6: Using keras, build and train a simple CNN model on the MNIST dataset from scratch. Include code for module creation, compilation, training, and evaluation.
**Answer:**

```python

# Importing libraries and loading dataset

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape data to include channel dimension (grayscale → 1 channel)
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

#Building the CNN Model

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')  # 10 classes for digits 0–9
])

print(model.summary())

# Compile the model

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Training the model

history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.1,
    verbose=2
)

# Evaluate the model

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")
```

Explanation
- Conv2D layers: Extract spatial features like edges and shapes from digit images.
- MaxPooling2D: Reduces spatial dimensions, making computation efficient while retaining important features.
- Flatten + Dense layers: Convert extracted features into a classification decision.
- Dropout: Prevents overfitting by randomly disabling neurons during training.
- Softmax output: Produces probability distribution across 10 digit classes (0-9).

# Question 7: Load and preprocess the CIFAR-10 dataset using Keras, and create a CNN model to classify RGB images. Show your preprocessing and architecture.
**Answer:**


```python
# Importing libraries and loading the dataset

import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize pixel values to range [0,1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Convert labels to one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)


# Building the CNN Model

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')  # 10 classes in CIFAR-10
])

print(model.summary())


# Compiling the model

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)


# Training the model

history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1,
    verbose=2
)

# Evaluating the model

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")
```

Explanation
- Preprocessing:
- Normalization scales pixel values to [0,1] for faster convergence.
- Labels are one-hot encoded since CIFAR-10 has 10 categories (airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck).
- Architecture:
- Three convolutional + pooling blocks progressively extract features from RGB images.
- Dense layers at the end perform classification.
- Dropout reduces overfitting.
- Softmax output layer provides probability distribution across 10 classes.

# Question 8: Using PyTorch, write a script to define and train a CNN on the MNIST dataset. Include model definition, data loaders, training loop, and accuracy evaluation.
**Answer:**

```python
# Importing libraries and loading dataset
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Transform: normalize pixel values to [0,1] and add channel dimension
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))  # mean and std for MNIST
])

# Load MNIST dataset
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# Data loaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1000, shuffle=False)


# Defining a CNN Model
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)   # 1 input channel (grayscale), 32 filters
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)  # 64 filters
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(64 * 12 * 12, 128)        # corrected flatten size
        self.fc2 = nn.Linear(128, 10)                  # 10 classes (digits 0–9)

    def forward(self, x):
        x = F.relu(self.conv1(x))          # -> (batch, 32, 26, 26)
        x = self.pool(F.relu(self.conv2(x)))  # -> (batch, 64, 12, 12)
        x = x.view(x.size(0), -1)          # flatten -> (batch, 64*12*12)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = CNN()

# Defining optimizer and loss function
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()


# Training Loop
epochs = 5
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f"Epoch {epoch+1}, Training Loss: {running_loss/len(train_loader):.4f}")

# Evaluation
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for data, target in test_loader:
        output = model(data)
        _, predicted = torch.max(output.data, 1)
        total += target.size(0)
        correct += (predicted == target).sum().item()

print(f"Test Accuracy: {100 * correct / total:.2f}%")
```

Explanation
- Data Loaders: Efficiently batch and shuffle MNIST data.
- CNN Architecture: Two convolutional layers with ReLU + pooling, followed by fully connected layers.
- Training Loop: Standard PyTorch loop with forward pass, loss computation, backpropagation, and optimizer step.
- Evaluation: Accuracy is computed on the test set using torch.max for predictions.

```python
# Creating dataset structure for training and inference

import zipfile
import os

zip_path = "cats-and-dogs-mini-dataset.zip"
extract_path = "dataset"

# Create the output directory if it doesn't exist
os.makedirs(extract_path, exist_ok=True)

# Unzip using Python
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Extraction complete.")


import os

# Create the 'data' directory if it doesn't exist
data_dir = 'data'
os.makedirs(data_dir, exist_ok=True)

# Create the 'train' and 'test' subdirectories within 'data'
train_dir = os.path.join(data_dir, 'train')
test_dir = os.path.join(data_dir, 'test')
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)


import os

# Define the source and destination paths
source_cat = 'dataset/cats_set'
dest_cat = 'dataset/cat'
source_dog = 'dataset/dogs_set'
dest_dog = 'dataset/dog'

# Rename the directories if they exist
if os.path.exists(source_cat):
  os.rename(source_cat, dest_cat)

if os.path.exists(source_dog):
  os.rename(source_dog, dest_dog)


# Creating the TRAIN TEST SPLIT within the DATA directory folders

import random
import shutil

def train_test_split_folder(source_folder, train_folder, test_folder, split_ratio=0.8):
    """
    Splits a folder of images into training and testing sets.

    Args:
        source_folder: Path to the source folder containing subfolders for each class.
        train_folder: Path to the folder where the training set will be saved.
        test_folder: Path to the folder where the testing set will be saved.
        split_ratio: The ratio of images to include in the training set (default is 0.8).
    """

    if not os.path.exists(train_folder):
        os.makedirs(train_folder)
    if not os.path.exists(test_folder):
        os.makedirs(test_folder)

    for class_name in os.listdir(source_folder):
        class_source_path = os.path.join(source_folder, class_name)

        if os.path.isdir(class_source_path):  # Check if it is a directory
            train_class_path = os.path.join(train_folder, class_name)
            test_class_path = os.path.join(test_folder, class_name)

            if not os.path.exists(train_class_path):
                os.makedirs(train_class_path)
            if not os.path.exists(test_class_path):
                os.makedirs(test_class_path)

            images = [f for f in os.listdir(class_source_path) if os.path.isfile(os.path.join(class_source_path, f))]
            random.shuffle(images)
            split_index = int(len(images) * split_ratio)
            train_images = images[:split_index]
            test_images = images[split_index:]

            for image in train_images:
                source_path = os.path.join(class_source_path, image)
                destination_path = os.path.join(train_class_path, image)
                shutil.copy(source_path, destination_path)

            for image in test_images:
                source_path = os.path.join(class_source_path, image)
                destination_path = os.path.join(test_class_path, image)
                shutil.copy(source_path, destination_path)


# Example usage (assuming you have your data organized in a 'data/train' folder):
train_test_split_folder("dataset", "data/train", "data/test")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam


# Preprocessing with ImageDataGenerator

image_size, batch_size = (128, 128), 32
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=20,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(train_dir, target_size=image_size,
                                               batch_size=batch_size, class_mode='binary')
test_data = test_datagen.flow_from_directory(test_dir, target_size=image_size,
                                             batch_size=batch_size, class_mode='binary')

# Build CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # binary classification
])
print(model.summary())

# Compile
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Training
history = model.fit(train_data, epochs=10, validation_data=test_data, verbose=2)

# Evaluation
test_loss, test_acc = model.evaluate(test_data, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")
```
