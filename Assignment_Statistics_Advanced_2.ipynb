Question 1: What is hypothesis testing in statistics?

Answer:

*  Hypothesis testing in statistics is a formal method used to decide whether data collected from a sample provides enough evidence to accept or reject a specific claim (the hypothesis) about a population parameter, such as a mean or proportion.
*  The Steps of Hypothesis Testing:

    * The process begins with stating the null and alternative hypotheses.

    * The significance level, usually 0.05 (5%), is chosen to control the likelihood of falsely rejecting the null hypothesis (Type I error).

    * Sample data is collected and an appropriate statistical test (such as t-test, z-test, or chi-square) is selected.

    * A test statistic is calculated, and its result is compared to a critical value or a p-value derived from the sample.

    * If the p-value is less than or equal to significance level, the null hypothesis is rejected in favor of the alternative. Otherwise, there isn't enough evidence to reject the null hypothesis.
*   Below is the python implementation of hypothesis testing for both a typical mean test and for the coin fairness example.

```python
import numpy as np
from scipy import stats

# Sample data: weights of 10-year-olds (kg)
data = np.array([34, 30, 35, 37, 32, 29, 33, 36, 31, 38])

# Null hypothesis: mean is 32kg
t_stat, p_value = stats.ttest_1samp(data, 32)
print("Test statistic:", t_stat)
print("p-value:", p_value)

# Decision
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

Question 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?

Answer:

*   The null hypothesis (H0) is a statement that there is no effect, no difference, or no relationship between variables in the population. It represents the default or status quo assumption that any observed effect is due to chance.
*   The alternative hypothesis (H1) is the statement that there is a genuine effect, difference, or relationship in the population. It is what the researcher aims to prove, opposing the null hypothesis.
*   Key Differences:

    * Assumption: The null hypothesis assumes no effect or difference, while the alternative hypothesis assumes there is an effect or difference.
    * Purpose: The null hypothesis is tested to be rejected or failed to be rejected, while the alternative hypothesis is what researchers aim to support with evidence.
    * Example: If testing a new drug's effectiveness, H0 might be the drug has no effect on recovery time, and H1 might be the drug reduces recovery time.
*   The below python code tests whether the average height of students is 5.5 feet or not:

```python
import numpy as np
from scipy import stats

# Sample data (in feet)
heights = np.array([5.1, 5.8, 5.6, 5.5, 5.7, 5.3, 5.4, 5.9, 5.2, 5.6])

# Null hypothesis (H0): mean height = 5.5 feet
# Alternative hypothesis (H1): mean height != 5.5 feet

t_stat, p_value = stats.ttest_1samp(heights, 5.5)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is a significant difference from 5.5 feet.")
else:
    print("Fail to reject null hypothesis: No significant difference from 5.5 feet.")
```

Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

Answer:

*   The significance level is the threshold probability set by the researcher before conducting a hypothesis test. It defines the maximum risk of rejecting the null hypothesis when it is actually true (a Type I error).
*   Role of Significance Level:

    * It controls how strong the evidence must be against the null hypothesis to make a rejection.

    * Typical values for significance level are 0.05, 0.01, or 0.10, with 0.05 (5%) being the most common.

    * If the p-value obtained from the test is less than or equal to significance level the null hypothesis is rejected, indicating statistically significant results.

    * If the p-value is greater than significance level, there is insufficient evidence to reject the null hypothesis, and the null is retained.

*   Significance Level vs. P-value:

    * The significance level is predetermined, while the p-value is computed from sample data.

    * Comparing p-value to significance level guides the decision to reject or fail to reject the null hypothesis.

*   The below python code tests if a sample mean significantly differs from 10, with a significance level of 0.05:

```python
import numpy as np
from scipy import stats

# Sample data
data = np.array([10.5, 9.8, 10.2, 9.7, 10.6, 10.1, 9.9, 10.3, 9.6, 10.4])

# Null hypothesis: population mean = 10
# Alternative hypothesis: population mean != 10

alpha = 0.05

t_stat, p_value = stats.ttest_1samp(data, 10)

print(f"Computed mean: {data.mean()}")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value <= alpha:
    print(f"Reject the null hypothesis at significance level {alpha}.")
else:
    print(f"Fail to reject the null hypothesis at significance level {alpha}.")
```

Question 4: What are Type I and Type II errors? Give examples of each.

Answer:

*   Type I and Type II errors are two types of errors that can be made in hypothesis testing when making decisions based on sample data.

    Type I Error (False Positive):
    * A Type I error occurs when the null hypothesis (H0) is true, but it is incorrectly rejected. This means concluding that there is an effect or difference when in reality there isn't one. The probability of committing a Type I error is equal to the significance level

    * Example: A medical test concludes a patient has a disease when they actually do not.

    Type II Error (False Negative):
    * A Type II error occurs when the null hypothesis is false, but it is incorrectly not rejected. This means failing to detect a real effect or difference. The probability of committing a Type II error is denoted by beta.
    * Example: A medical test fails to detect a disease in a patient who actually has it.

*   Python Code Illustration for Type I and Type II Errors:

```python
import numpy as np
from scipy import stats

np.random.seed(42)

# True population mean under null hypothesis
mu_0 = 50

# Sample from population with true mean 50 (null true)
sample_null_true = np.random.normal(mu_0, 5, 100)

# Sample from population with true mean 55 (null false)
sample_null_false = np.random.normal(55, 5, 100)

alpha = 0.05

# Type I error simulation: reject H0 when true
t_stat_null_true, p_val_null_true = stats.ttest_1samp(sample_null_true, mu_0)
print("Type I error example:")
print(f"P-value: {p_val_null_true}")
if p_val_null_true < alpha:
    print("Reject null hypothesis (Type I error if null is true).")
else:
    print("Fail to reject null hypothesis (correct behavior).")

# Type II error simulation: fail to reject H0 when false
t_stat_null_false, p_val_null_false = stats.ttest_1samp(sample_null_false, mu_0)
print("\nType II error example:")
print(f"P-value: {p_val_null_false}")
if p_val_null_false > alpha:
    print("Fail to reject null hypothesis (Type II error if null is false).")
else:
    print("Reject null hypothesis (correct behavior).")
```

Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.

Answer:

*   Difference betweeen a Z-test and T-test:
    * The Z-test is used when we have a large sample size, while T-test is used when we have a small sample size.
    * The Z-test is used when the population standard deviation is known, while T-test is used when population standard deviation is unknown.
    * The Example of Z-test:

      A school principal believes their students have a higher IQ than the state average (population mean = 100, population std = 20). They take a sample of 50 students and want to test if their sample mean IQ is significantly different from 100. Since the sample size is large and the population std is known, a Z-test is appropriate.

    * The EXample of T-test:

      Suppose we want to test if the average weight of a type of apple differs from 150 grams. If we take a small random sample of 15 apples and find the sample mean and sample std dev. Since the population std is unknown and sample is small, use a T-test.
*   The below python Code Example Demonstrates both Z-test and T-test:

```python
import numpy as np
from scipy import stats

# Example for Z-test (large sample, known population std dev)
popmean = 100
popstd = 20
sample_size_large = 50
sample_large = np.random.normal(110, popstd, sample_size_large)

# Z-test statistic calculation
z_stat = (np.mean(sample_large) - popmean) / (popstd / np.sqrt(sample_size_large))
p_value_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))
print(f"Z-test statistic: {z_stat:.3f}, p-value: {p_value_z:.3f}")

# Example for T-test (small sample, unknown population std dev)
popmean = 150
sample_small = np.array([160, 158, 152, 149, 155, 162, 147, 151, 159, 148, 153, 157, 150, 154, 149])

t_stat, p_value_t = stats.ttest_1samp(sample_small, popmean)
print(f"T-test statistic: {t_stat:.3f}, p-value: {p_value_t:.3f}")
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

Answer:

```python
import numpy as np
from scipy import stats

# Sample data given
sample_data =  [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population mean (hypothetical)
population_mean = 50

# Known population standard deviation (assuming known for Z-test)
# Estimating std from sample data for demonstration (usually should be known)
population_std = np.std(sample_data, ddof=0)

# Sample size and sample mean
n = len(sample_data)
sample_mean = np.mean(sample_data)

# Calculate Z statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate two-tailed p-value from standard normal distribution
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject the null hypothesis: There is significant evidence the mean differs from 50.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the mean differs from 50.")
```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(42)

# Simulate data from a normal distribution
mu = 50  # True mean
sigma = 5  # True standard deviation
n = 100  # Sample size
data = np.random.normal(mu, sigma, n)

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
standard_error = sample_std / np.sqrt(n)

# Calculate 95% confidence interval
z_critical = norm.ppf(0.975)  # Z-score for 95% CI (two-tailed)
margin_of_error = z_critical * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, density=True, alpha=0.7, color='skyblue', label='Sample Data')
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Sample Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='--', label=f'95% CI Lower = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'95% CI Upper = {ci_upper:.2f}')

# Plot the true normal distribution
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma), 'k-', lw=2, label='True Normal Distribution')
plt.title('Simulated Normal Distribution with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

# Below is the python function that calculates the z-scores from sample dataset and visualize the
# standardized data using a histogram
def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot a histogram of the standardized data.

    Parameters:
        data (list or np.array): Input dataset
    """
    # Convert data to numpy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Sample standard deviation

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Print summary statistics
    print(f"Mean of original data: {mean:.2f}")
    print(f"Standard deviation of original data: {std:.2f}")
    print(f"Mean of Z-scores: {np.mean(z_scores):.2f}")
    print(f"Standard deviation of Z-scores: {np.std(z_scores, ddof=1):.2f}")

    # Plot histogram of Z-scores
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=20, density=True, alpha=0.7, color='skyblue', label='Z-scores')
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z = 0)')
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Density')

    # Overlay standard normal distribution
    x = np.linspace(-4, 4, 100)
    plt.plot(x, 1/(np.sqrt(2*np.pi)) * np.exp(-x**2/2), 'k-', lw=2,
             label='Standard Normal Distribution')

    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage with sample data
if __name__ == "__main__":
    # Sample dataset
    sample_data = [49.1, 50.2, 51.0, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                   50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                   50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

    z_scores = calculate_and_plot_z_scores(sample_data)
```
