## Question 2. How does K-Means clustering algorithm work?

Answer:

*   K-Means is one of the simplest and most popular unsupervised machine learning algorithms used for partitioning a dataset into K distinct, non-overlapping clusters. The goal is to group similar data points together and separate dissimilar ones, such that points in the same cluster are as close as possible to each other, while clusters are as far apart as possible.
The "means" in K-Means refers to the centroid (average position) of each cluster.

*   How K-Means Works (Step-by-Step)

    Choose the number of clusters K

    * You must specify K in advance (e.g., K=3). Choosing the right K is a separate problem (often solved with the Elbow method, silhouette score, etc.).

    Initialize K centroids

    Randomly place K points in the data space as initial cluster centers. Common methods:

    * Randomly select K actual data points
    
    * Or place them randomly in the feature space.(Better initialization like K-Means++ is usually used in practice to avoid bad starting positions.)

    Assignment step (repeat until convergence)
    
    For every data point:

    * Calculate its distance (usually Euclidean distance) to all K centroids.

    * Assign the point to the cluster whose centroid is closest.

    Update step (re-calculate centroids)

    For each cluster:
    
    * Compute the mean (average) of all points now assigned to that cluster.

    * Move the centroid to this new mean position.

    Repeat steps 3 and 4

    Keep alternating between assignment and update until:

    * The centroids no longer move (or move less than a tiny threshold), or

    * The assignments no longer change, or

    * A maximum number of iterations is reached.

    Output

    * Final K centroids

    * Assignment of every data point to one of the K clusters

## Question 3. Explain the concept of a dendrogram in hierarchical clustering.

Answer:

*   A dendrogram in hierarchical clustering is a tree-like diagram that visually represents the arrangement and merging of clusters. It shows how individual data points or clusters are progressively combined (or split) based on their similarity or distance. The vertical axis of the dendrogram represents the distance or dissimilarity between clusters at the point they merge. Lower merge points indicate higher similarity between the joined clusters, while higher points show greater dissimilarity.

*   Each leaf or endpoint at the bottom of a dendrogram corresponds to a single data point. As you move upward, clusters are joined together step-by-step, showing the hierarchy of nested clusters from individual points up to one large cluster containing all data. By "cutting" the dendrogram horizontally at a certain height, you can determine the number of clusters and which points belong to each cluster. The height at which you cut depends on the cluster distance or how granular you want the clustering to be.

*   In summary, a dendrogram is a core visualization tool in hierarchical clustering that helps understand the cluster formation process, the relationships between data points at different levels, and how to select the appropriate number of clusters for your analysis.

## Question 4. What is the main difference between K-Means and Hierarchical Clustering?

Answer:

The main difference between K-Means and Hierarchical Clustering lies in their approach to forming clusters and how clusters are defined:

*   K-Means clustering is a partitional method that requires the number of clusters (K) to be specified beforehand. It partitions the data into K non-overlapping, spherical clusters by iteratively assigning points to the nearest cluster centroid and updating centroids until convergence. It is computationally efficient and suited for large datasets with clusters of roughly equal size and shape.

*   Hierarchical clustering does not need a predefined number of clusters. Instead, it builds a tree-like structure (dendrogram) by either progressively merging individual points into clusters (agglomerative) or splitting a single cluster into smaller clusters (divisive). This method can capture clusters of varying shapes and sizes and provides a visual hierarchy of clusters. However, it is more computationally intensive and better suited for smaller datasets.

## Question 5. What are the advantages of DBSCAN over K-Means?

Answer:

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) offers several advantages over K-Means clustering:

*   DBSCAN can find clusters of arbitrary shapes, whereas K-Means tends to find spherical clusters due to its reliance on centroids.

*   It does not require specifying the number of clusters beforehand, unlike K-Means which requires the number k of clusters as input.

*   DBSCAN is capable of identifying noise and outliers, labeling them as points that do not belong to any cluster. K-Means assigns all points to clusters, which can distort cluster centers if outliers are present.

*   DBSCAN works well with datasets containing clusters of varying sizes and densities, while K-Means assumes clusters of similar size and density.

*   It is less sensitive to initialization and does not require random seeding, which makes DBSCAN results more stable compared to K-Means.

*   These advantages make DBSCAN particularly suitable for spatial or irregularly shaped data and for applications where the number of clusters is unknown or outliers need to be detected effectively.

## Question 6.  When would you use Silhouette Score in clustering?

Answer:

*   The Silhouette Score is used in clustering to evaluate how well the data points have been clustered. It measures both cohesion (how close a point is to other points in the same cluster) and separation (how distinct or far a point is from points in other clusters). The score ranges from -1 to 1, where a value close to +1 indicates that points are well matched to their own cluster and poorly matched to neighboring clusters, signaling well-defined clusters. A score near 0 suggests overlapping clusters, and negative values indicate potential misclassification of points.

*   You would use the Silhouette Score mainly to:

    * Assess the quality of clustering results after performing clustering.

    * Compare different clustering algorithms or settings to select the best model.

    * Determine the optimal number of clusters by evaluating Silhouette Scores across different cluster counts.

    * Detect anomalies or misclassified points by identifying negative silhouette values.

*   Overall, it's a valuable internal validation metric to help ensure clusters are cohesive and well-separated, especially when true labels are unavailable in unsupervised clustering.

## Question 7.  What are the limitations of Hierarchical Clustering?

Answer:

The limitations of hierarchical clustering include:

*   Computational complexity: Hierarchical clustering can be very slow and computationally expensive, especially for large datasets, as it evaluates distances between all pairs of points or clusters.

*   Sensitivity to outliers: Outliers can significantly impact the resulting clusters and distort the dendrogram structure.

*   Difficulty in choosing linkage and distance metrics: The results heavily depend on the choice of linkage criteria (e.g., single, complete, average) and distance metrics, which are often chosen arbitrarily and can lead to different clustering outcomes.

*   No automatic determination of clusters: Unlike some algorithms, hierarchical clustering doesn't automatically provide an optimal number of clusters; manual cutting of the dendrogram or external metrics is required.

*   Inflexibility in undoing merges: Once clusters are merged or split at a step, those decisions cannot be reversed, which may lead to suboptimal final clusters.

*   Poor handling of large or high-dimensional datasets: Due to computational demands and memory usage, it is less suitable for very large or high-dimensional data.

*   Issues with mixed data types and missing values: It struggles with datasets that have both numerical and categorical variables or that contain missing data without preprocessing.

*   These limitations make hierarchical clustering more appropriate for small to moderate-sized datasets where the dendrogram's hierarchical structure provides valuable insights, rather than large-scale or complex data scenarios.

## Question 8.  Why is feature scaling important in clustering algorithms like K-Means?

Answer:

*   Feature scaling is important in clustering algorithms like K-Means because the algorithm relies on distance metrics (usually Euclidean distance) to measure similarity between data points. If features have different scales or units, features with larger ranges will disproportionately influence the distance calculations, causing the clustering process to be biased toward those features. This can lead to poor clustering results, where clusters are dominated by high-magnitude features rather than meaningful patterns.

*   Scaling features to a common range or distribution ensures that each feature contributes equally to the distance computation, improving the accuracy and meaningfulness of the clusters. Additionally, feature scaling helps the K-Means algorithm converge faster to a stable solution.

*   In summary, feature scaling prevents dominance of variables with larger scales, balances feature importance, improves clustering performance, and speeds up convergence in K-Means and other distance-based clustering algorithms.

## Question 9. How does DBSCAN identify noise points?

Answer:

*   DBSCAN identifies noise points based on density criteria defined by two parameters: epsilon (ε), which is the radius around a point to consider its neighbors, and minPts, the minimum number of points required within that radius to form a dense region.

*   A point is classified as noise if it is neither a core point (having at least minPts neighbors in its ε-radius) nor a border point (within ε-distance of a core point but with fewer neighbors itself).

*   Essentially, noise points are isolated, lying in low-density regions where they do not satisfy the density requirements to belong to any cluster. These points remain unassigned to any cluster and are treated as outliers by the algorithm.

## Question 10. Define inertia in the context of K-Means.

Answer:

*   In the context of K-Means clustering, inertia is defined as the sum of squared distances between each data point and the centroid of the cluster to which it belongs. It is also called the within-cluster sum-of-squares (WCSS).

*   Mathematically, inertia measures how internally coherent the clusters are by calculating the total squared distance of points from their respective cluster centers. Lower inertia values indicate more compact and well-defined clusters, meaning the data points are closer to their cluster centroids.

*   Inertia is commonly used as a metric to evaluate the quality of clustering, and it is the quantity that K-Means tries to minimize during the clustering process. However, inertia is not normalized and can be influenced by the number of clusters and dimensionality of the data.

*   In summary, inertia quantifies the tightness of clusters formed by K-Means and serves as a criterion for optimizing the clustering performance.

## Question 11. What is the elbow method in K-Means clustering?

Answer:

*   The elbow method in K-Means clustering is a graphical technique used to determine the optimal number of clusters (K) to use. It involves plotting the within-cluster sum of squares (WCSS), also known as inertia, against different values of K. The WCSS measures how compact the clusters are by summing the squared distances between points and their respective cluster centroids.

*   As we increase K, WCSS decreases because having more clusters typically means tighter, more specific groupings. The elbow method looks for a point on the plot where the rate of decrease sharply changes, forming an "elbow" shape. This point suggests that adding more clusters beyond this number yields diminishing returns in terms of improving cluster compactness. Therefore, the K at this elbow is considered an appropriate balance between cluster cohesion and complexity.

* In practice, the elbow method helps choose a K value that avoids underfitting (too few clusters) or overfitting (too many clusters), providing a good trade-off for clustering quality and interpretability.

## Question 12.  Describe the concept of "density" in DBSCAN.

Answer:

*   Density in DBSCAN refers to how closely packed the data points are in a given region of the data space. It is defined based on two parameters: epsilon (ε), which sets the radius to consider neighbors around a point, and MinPts, the minimum number of points required within this radius to declare a region as dense.

*   A dense region is one where a point has at least MinPts neighbors within its ε-radius, classifying it as a core point. Clusters are then formed by connecting core points and their neighbors that meet this density requirement. Points in sparse regions, where density falls below this threshold, are treated as noise or outliers.

*   This concept of density allows DBSCAN to identify clusters of arbitrary shape and distinguish noise without needing the number of clusters upfront.

## Question 15.  Explain the term "linkage criteria" in hierarchical clustering.

Answer:

*   Linkage criteria in hierarchical clustering refer to the rule or strategy used to calculate the distance or dissimilarity between clusters when deciding which clusters to merge or split. Since hierarchical clustering works by iteratively combining or dividing clusters, the linkage method determines how the distance between sets of points (clusters) is measured based on distances between individual points within those clusters.

*   Common linkage methods include:

    Single linkage:
    
    * Distance between two clusters is the minimum distance between any pair of points from each cluster. It tends to create long, chain-like clusters.

    Complete linkage:
    
    * Distance is the maximum distance between any pair of points from each cluster, creating more compact and spherical clusters.

    Average linkage:
    
    * Distance is the average of all pairwise distances between points in the two clusters, providing a balance between single and complete linkage.

    Ward's linkage:
    
    * Minimizes the increase in total within-cluster variance after merging clusters, promoting compact, spherical clusters.

*   Choosing the right linkage criterion affects the shape, cohesion, and separation of clusters and influences the final dendrogram structure. Thus, linkage criteria are fundamental in hierarchical clustering for determining how clusters are formed and merged.

## Question 16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Answer:

*   K-Means clustering tends to perform poorly on data with varying cluster sizes or densities because it assumes clusters to be roughly spherical and of similar size. Specifically, K-Means assigns points based on proximity to cluster centroids using Euclidean distance, which causes several challenges:

    * Clusters with different sizes or densities can distort centroid placement, causing smaller or less dense clusters to be merged incorrectly or overshadowed by larger, denser ones.

    * K-Means minimizes within-cluster variance assuming relatively uniform cluster scale, so clusters with widely differing densities violate this assumption.

    * The algorithm is biased toward equal-sized, globular clusters, failing to identify clusters with irregular shapes or variable density.

    * Outliers or noise can disproportionately influence centroid calculation, skewing clusters further in such uneven settings.

*   This limitation leads to poor representation of true underlying cluster structure in datasets with clusters of different volumes or densities. Alternative algorithms like DBSCAN or hierarchical clustering, which do not rely on such assumptions, are better suited for these cases.

## Question 17. What are the core parameters in DBSCAN, and how do they influence clustering?

Answer:

*   The core parameters in DBSCAN are:

    Epsilon (ε):
    
    * This defines the maximum radius or distance within which points are considered neighbors. It determines the neighborhood size for density estimation. A smaller ε results in smaller, tighter clusters, while a larger ε may merge distinct clusters into one.

    MinPts (Minimum Points):
    
    * This specifies the minimum number of points required within the ε-radius neighborhood to form a dense region, classifying a point as a core point. Higher MinPts leads to denser clusters and more points labeled as noise, while lower MinPts may result in larger and less dense clusters.

*   These parameters influence clustering by controlling how DBSCAN detects dense areas and forms clusters:

    Points with at least MinPts neighbors within ε are core points, starting cluster formation.

    Points within ε of core points but not meeting MinPts themselves are border points, belonging to clusters but not cores.

    Points outside ε neighborhoods of core points and not meeting MinPts are labeled as noise or outliers.

*   Proper tuning of ε and MinPts is essential for DBSCAN's effectiveness, as they dictate cluster size, shape, and sensitivity to noise. Selecting these parameters depends on the specific dataset and domain knowledge.

## Question 18. How does K-Means++ improve upon standard K-Means initialization?

Answer:

*   K-Means++ improves upon standard K-Means initialization by selecting initial centroids in a smarter way, which leads to better clustering results and faster convergence. Unlike the standard K-Means, which randomly picks initial centroids, K-Means++ distributes the initial centroids across the data space according to a probability proportional to the squared distance from existing centroids.

*   The process works as follows:

    * The first centroid is chosen uniformly at random from the data points.

    * For each subsequent centroid, points are selected with probability proportional to the squared distance from the nearest already chosen centroid.

    * This spreading ensures that initial centroids are well-separated, reducing the chances of poor local optima.

*   By initializing more strategically, K-Means++ typically results in:

    * Better cluster separation

    * More consistent and stable results

    * Faster convergence to an optimal clustering

*   Although this method incurs slightly more computation during initialization, it generally reduces overall runtime by requiring fewer iterations for convergence and improves the quality of the clustering.

## Question 19. What is agglomerative clustering?

Answer:

*   Agglomerative clustering is a type of hierarchical clustering that works by successively merging the most similar (or closest) pairs of clusters, starting from individual data points. It is a bottom-up approach where initially each data point is considered as its own cluster, and then these clusters are gradually combined until all points are grouped into a single large cluster or until a stopping criterion is reached.

*   How it works:

    Start:
    
    * Each data point is treated as a single-cluster.

    Iterative merging:
    
    * The algorithm calculates the similarity (or distance) between all pairs of clusters using a linkage criterion.

    Merge:
    
    * The two most similar clusters are merged into a new, larger cluster.

    Repeat:
    
    * This process continues, repeatedly merging the closest pairs of clusters based on the chosen linkage method, until only one cluster remains or a predefined number of clusters is reached.

*   Key point:

    The result can be visualized as a dendrogram, a tree-like diagram illustrating the order in which clusters are merged and the similarity at each step. This provides insight into the data's structure, such as how clusters relate to each other at different levels of similarity or dissimilarity.

*   Linkage criteria:

    Different methods can be used to measure the similarity between clusters, such as single linkage (minimum distance), complete linkage (maximum distance), average linkage, or Ward's method (minimizing variance).

*   Agglomerative clustering is a flexible method for discovering hierarchical relationships in data that can be visualized and analyzed at various levels of granularity.

## Question 20.  What makes Silhouette Score a better metric than just inertia for model evaluation?

Answer:

*   The Silhouette Score is considered a better metric than just inertia for model evaluation because it provides a more nuanced and comprehensive assessment of clustering quality.

*   Key differences:

    * Inertia measures the compactness of clusters by summing squared distances of points to their cluster centers. While lower inertia indicates tighter clusters, it only focuses on how internally cohesive the clusters are and tends to decrease as the number of clusters increases, sometimes misleadingly suggesting overfitting.

    * Silhouette Score evaluates both cohesion (how close points are within the same cluster) and separation (how far points are from points in the next nearest cluster). It ranges from -1 to 1, with higher values indicating better-defined, well-separated clusters.

*   Why Silhouette Score is better:


    Balanced evaluation:
    
    * It considers both how compact the clusters are and how distinct they are from each other, providing a holistic view of clustering quality.

    Scale independence:
    
    * Unlike inertia, the Silhouette Score is normalized and not affected by the scale or size of the dataset or the number of features.

    Robustness to overfitting:
    
    * It discourages the need to overcluster, which might artificially decrease inertia, but can lead to meaningless clusters.

    Optimal number of clusters:
    
    * It is particularly useful for determining the best number of clusters in an unsupervised setting because it directly assesses the separation quality.

## Question 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

Answer:


```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.8, marker='X', label='Centroids')
plt.title('K-Means Clustering of Synthetic Data with 4 Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

## Question 23.  Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# 1. Generate synthetic moon-shaped data
X, y = make_moons(n_samples=300, noise=0.08, random_state=42)

# 2. Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# 3. Identify outliers (label = -1)
outliers = labels == -1
clusters = labels >= 0

# 4. Plot results
plt.figure(figsize=(8, 6))

# Plot clustered points
plt.scatter(
    X[clusters, 0],
    X[clusters, 1],
    c=labels[clusters],
    s=40,
    edgecolor='k',
    alpha=0.8
)

# Plot outliers
plt.scatter(
    X[outliers, 0],
    X[outliers, 1],
    color='red',
    marker='x',
    s=100,
    linewidths=2,
    label='Outliers'
)

plt.title("DBSCAN on make_moons — Outliers Highlighted")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.tight_layout()
plt.show()
```

## Question 24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.

Answer.

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# 1. Load Wine dataset
data = load_wine()
X = data.data

# 2. Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# 4. Print size of each cluster
unique, counts = np.unique(labels, return_counts=True)

print("Cluster sizes:")
for cluster_id, size in zip(unique, counts):
    print(f"Cluster {cluster_id}: {size} samples")
```

Answer:

```python
"""
Generate synthetic data with sklearn.make_blobs using varying cluster standard deviations
and cluster it using DBSCAN. The script prints cluster sizes and plots core vs border vs noise.
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from collections import Counter


# 1) Generate synthetic data
random_state = 42
centers = [(-5, -2), (0, 0), (4, 4), (8, 0)]            # 4 cluster centers
cluster_std = [0.3, 0.9, 0.5, 1.5]                      # varying standard deviations
n_samples = [150, 250, 180, 120]                        # samples per cluster

X, y_true = make_blobs(n_samples=n_samples,
                       centers=centers,
                       cluster_std=cluster_std,
                       random_state=random_state)

# Optionally scale features (often helps DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# 2) Run DBSCAN
# NOTE: eps/min_samples often need tuning depending on scale/noise.
eps = 0.5            # try changing this (e.g., 0.4, 0.6, 0.8) to see effect
min_samples = 8
db = DBSCAN(eps=eps, min_samples=min_samples)
labels = db.fit_predict(X_scaled)   # -1 = noise

# core / border masks
core_sample_mask = np.zeros_like(labels, dtype=bool)
if hasattr(db, "core_sample_indices_"):
    core_sample_mask[db.core_sample_indices_] = True
border_mask = (labels != -1) & (~core_sample_mask)
noise_mask = labels == -1


# 3) Print cluster sizes

counts = Counter(labels)
n_clusters = len([lab for lab in counts.keys() if lab != -1])
n_noise = counts.get(-1, 0)

print(f"DBSCAN parameters: eps={eps}, min_samples={min_samples}")
print(f"Estimated number of clusters (excluding noise): {n_clusters}")
print(f"Number of noise points (label -1): {n_noise}")
print("Cluster sizes (label: count):")
for lab, cnt in sorted(counts.items()):
    print(f"  {lab}: {cnt}")

# -------------------------
# 4) Plot results
# -------------------------
plt.figure(figsize=(9, 7))

# colors for clusters (ignore -1)
unique_labels = sorted([lab for lab in set(labels) if lab != -1])
n_unique = len(unique_labels)
cmap = plt.get_cmap("tab10")

# Plot core points (larger markers)
for i, lab in enumerate(unique_labels):
    mask = (labels == lab) & core_sample_mask
    plt.scatter(X_scaled[mask, 0],
                X_scaled[mask, 1],
                s=80,
                marker='o',
                edgecolor='k',
                label=f'Cluster {lab} (core)',
                alpha=0.9,
                color=cmap(i % 10))

# Plot border points (smaller markers)
for i, lab in enumerate(unique_labels):
    mask = (labels == lab) & border_mask
    plt.scatter(X_scaled[mask, 0],
                X_scaled[mask, 1],
                s=30,
                marker='o',
                edgecolor='k',
                label=f'Cluster {lab} (border)' if np.any(mask) else None,
                alpha=0.7,
                color=cmap(i % 10))

# Plot noise (outliers)
if np.any(noise_mask):
    plt.scatter(X_scaled[noise_mask, 0],
                X_scaled[noise_mask, 1],
                s=100,
                marker='x',
                linewidths=2,
                color='black',
                label='Noise (outliers)')

plt.title("DBSCAN clustering on make_blobs (varying stds) — scaled features")
plt.xlabel("Feature 1 (scaled)")
plt.ylabel("Feature 2 (scaled)")
plt.legend(loc='best', fontsize='small', ncol=2)
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
"""
Load the Digits dataset, reduce to 2D with t-SNE, cluster with DBSCAN, and visualize results.

Run in a Jupyter notebook or standard Python script. t-SNE is stochastic; set random_state for
reproducibility. You may need to tune DBSCAN's eps and min_samples for different t-SNE runs.
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from collections import Counter


# 1. Load & preprocess

digits = load_digits()
X = digits.data         # shape (n_samples, n_features)
y_true = digits.target  # true digit labels (0-9)

# Standardize features before t-SNE (recommended)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# 2. t-SNE reduction

# Choose perplexity based on n_samples (default 30 is usually OK)
tsne = TSNE(n_components=2, perplexity=30, init='pca', random_state=42, learning_rate='auto')
X_tsne = tsne.fit_transform(X_scaled)   # shape (n_samples, 2)


# 3. DBSCAN on t-SNE embedding

# eps/min_samples often need tuning. Start with eps=2.0..3.0 when using standard t-SNE settings.
dbscan = DBSCAN(eps=2.5, min_samples=5)
labels = dbscan.fit_predict(X_tsne)  # -1 => noise

# Summary of clusters
counts = Counter(labels)
n_clusters = len([lbl for lbl in counts.keys() if lbl != -1])
n_noise = counts.get(-1, 0)
print(f"DBSCAN found {n_clusters} clusters (excluding noise) and {n_noise} noise points")
print("Cluster size distribution (label: count):")
for lbl, cnt in sorted(counts.items()):
    print(f"  {lbl:3d}: {cnt}")


# 4. Plot: t-SNE colored by DBSCAN labels

plt.figure(figsize=(10, 6))
# scatter colored by DBSCAN label; use a colormap that handles many labels
sc = plt.scatter(X_tsne[:, 0], X_tsne[:, 1],
                 c=labels, cmap='tab20', s=40, edgecolor='k', alpha=0.85)

# highlight noise points with black 'x'
noise_mask = labels == -1
if noise_mask.any():
    plt.scatter(X_tsne[noise_mask, 0], X_tsne[noise_mask, 1],
                marker='x', s=100, linewidths=2, color='black', label='Noise')

plt.title("Digits dataset: t-SNE (2D) + DBSCAN clustering")
plt.xlabel("t-SNE component 1")
plt.ylabel("t-SNE component 2")
plt.colorbar(sc, label='DBSCAN label (colors)')
plt.legend(loc='best')
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()


# 5. Plot: t-SNE colored by true digit labels (for comparison)

plt.figure(figsize=(10, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1],
            c=y_true, cmap='tab10', s=40, edgecolor='k', alpha=0.85)
plt.title("Digits dataset: t-SNE (2D) colored by true digit labels")
plt.xlabel("t-SNE component 1")
plt.ylabel("t-SNE component 2")
plt.colorbar(label='True digit label')
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import numpy as np

# Load Iris dataset
data = load_iris()
X = data.data

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(X)

# Compute Silhouette Scores for each sample
silhouette_vals = silhouette_samples(X, cluster_labels)

# Plot Silhouette Coefficient for each sample
plt.figure(figsize=(10, 6))
y_ticks = []
y_lower = 10

for i in range(3):
    ith_cluster_silhouette_values = silhouette_vals[cluster_labels == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      alpha=0.7)

    # Label the silhouette plots with their cluster numbers in the middle
    y_ticks.append((y_lower + y_upper) / 2)
    y_lower = y_upper + 10  # 10 for the 0 samples gap between clusters

plt.yticks(y_ticks, [f'Cluster {i}' for i in range(3)])
plt.xlabel('Silhouette coefficient values')
plt.ylabel('Cluster')
plt.title('Silhouette Coefficient for each sample in Iris dataset')
plt.show()
```

## Question 45. Generate synthetic data using make_blobs and apply Agglomerative Clustering with 'average' linkage. Visualize clusters.

Answer:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Apply Agglomerative Clustering with average linkage
agg_clustering = AgglomerativeClustering(n_clusters=4, linkage='average')
labels = agg_clustering.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('Agglomerative Clustering with Average Linkage')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Question 46.  Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).


Answer:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans

# Load the Wine dataset
wine = load_wine(as_frame=True)
df = wine.frame

# Use first 4 features for visualization
features = wine.feature_names[:4]
df_sub = df[features].copy()

# Apply KMeans clustering with 3 clusters (as Wine dataset has 3 classes)
kmeans = KMeans(n_clusters=3, random_state=42)
df_sub['cluster'] = kmeans.fit_predict(df_sub)

# Plot pairplot colored by cluster assignment
sns.pairplot(df_sub, hue='cluster', palette='viridis')
plt.suptitle('KMeans Clustering on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
```

## Question 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

Answer:

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate synthetic noisy blobs
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.2, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Identify clusters and noise
n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
n_noise = np.sum(dbscan_labels == -1)

print(f'Number of clusters found: {n_clusters}')
print(f'Number of noise points found: {n_noise}')
```

## Question 48.  Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

Answer:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

# Load Digits dataset
digits = load_digits()
X = digits.data

# Reduce dimensions using t-SNE to 2 components
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=10)
labels = agg_clustering.fit_predict(X_tsne)

# Plot clusters in 2D
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=50)
plt.title('Agglomerative Clustering on Digits Dataset with t-SNE reduction')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```
